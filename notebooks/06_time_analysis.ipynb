# 淘宝用户行为数据分析 - 时间维度分析

## 1. 加载数据

本Notebook用于分析用户行为的时间分布特征，包括时段、日期、周几等维度。

```python
import sys
import pandas as pd
sys.path.append('../src')

from data_loader import load_cleaned_data
from metrics import calculate_hourly_metrics,calculate_daily_metrics,calculate_weekly_metrics
from plot_utils import plot_hourly_metrics,plot_daily_metrics,plot_weekly_metrics

df = load_cleaned_data()
print(f'本次共加载{len(df)}条数据')
```

## 2.小时维度分析

```python
hourly_metrics = calculate_hourly_metrics(df)
display(hourly_metrics)
```

数据可视化

```python
plot_hourly_metrics(hourly_metrics)
```

## 3.日期维度分析

```python
daily_metrics = calculate_daily_metrics(df)
display(daily_metrics)
```

数据可视化

```python
plot_daily_metrics(daily_metrics)
```

## 4.周几维度分析

```python
df['time'] = pd.to_datetime(df['time'])
df['weekday'] = df['time'].dt.dayofweek
df['weekday_name'] = df['time'].dt.day_name()

weekly_metrics = calculate_weekly_metrics(df)
display(weekly_metrics)
```

数据可视化

```python
plot_weekly_metrics(weekly_metrics)
```
